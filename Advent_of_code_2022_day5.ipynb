```python
df[['move', 'from', 'to']] = df['moving'].str.split(' from | to ', expand=True)
```

```python
df['move'] = df['move'].str.replace('move ', '')
```

```python
df.head(3)
```

```python
def perform_moves(df, stacks):
    for index, row in df.iterrows():
        move_item = row['move']
        from_index = row['from']
        to_index = row['to']

        for i, stack in enumerate(stacks):
            from_index = int(from_index)
            to_index = int(to_index)
            if 0 <= from_index < len(stack):
                item_to_move = stack[from_index]
                stack.pop(from_index)
                stack.insert(to_index, item_to_move)
```

```python
perform_moves(df, stacks)
```
