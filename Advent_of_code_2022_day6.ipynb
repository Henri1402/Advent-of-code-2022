```python
def find_unique_substrings(data):
    unique_substrings = []
    for i in range(len(data)):
        substring = data[i:i+4]
        if len(set(substring)) == 4:
            unique_substrings.append(f"characters {i+1}-{i+4}")
    return unique_substrings
```
